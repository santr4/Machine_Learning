Generate test datasets using sklearn.datasets


```python
# ** 2D binary classification data generated by sklearn.datasets
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

x,y = make_circles(n_samples=200, shuffle=True,noise=0.1,random_state=42)
train_x,train_y = make_circles(n_samples=500, shuffle=True,noise=0,random_state=42)  # here nosie is 0 so perfect circles.
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
plt.scatter(train_x[:,0],train_x[:,1],c=train_y)
plt.show()
```

```python
# ** two interlocking half circles using make_moons function
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

x,y = make_moons(n_samples=500,shuffle=True,noise=0.15,random_state=42)
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
plt.clf()
```

```python
# ** make_blobs function is used for make clusters.
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

x,y = make_blobs(n_samples=500,centers=3,n_features=2,random_state=23)

plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
plt.clf()
```

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt

x,y = make_classification(n_samples=100,n_features=2,n_redundant=0,n_informative=2,n_repeated=0,n_classes=3,n_clusters_per_class=1)
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
# ** a random multilabel classification dataset is formed by make_multilabel_classificaton()
from sklearn.datasets import make_multilabel_classification
import pandas as pd
import matplotlib.pyplot as plt

x,y = make_multilabel_classification(n_samples=500,n_features=2,n_classes=2,n_labels=2,allow_unlabeled=True,random_state=23)

# now we have to create pandas dataframe from the fake generated dataset from sklearn
df = pd.concat([pd.DataFrame(x,columns=['x1','x2']),
                pd.DataFrame(y,columns=['Label1','Label2'])],
                axis=1)

display(df.head())
df.head() # both doing the same work

plt.scatter(df['x1'],df['x2'],c=df['Label1'])
plt.show()
plt.clf()
```

```python
# now we will target for linear regression using make_regression()
# this graph shows the linear regression.
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
x,y = make_regression(n_samples=50,n_features=1,noise=20,random_state=23)
plt.scatter(x,y)
plt.show()
plt.clf()
```
