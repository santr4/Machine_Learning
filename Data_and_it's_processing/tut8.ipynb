one-hot encoding in python

one-hot encoding with iris dataset


```python
import pandas as pd
```

```python
data = pd.read_csv("/home/aaranyak/Downloads/IRIS.csv")
data.head()
```

```python
# we can create dummy variables to store the species column
dummy = pd.get_dummies(data.species)
dummy.tail()
```

```python
# concat the dummy column to the dataset
merged_data = pd.concat([data,dummy],axis=1)
```

```python
merged_data.head()
```

```python
# now since we have done the one hot encoding we can drop the species column
final_data = merged_data.drop(columns='species')
final_data.head()
```
